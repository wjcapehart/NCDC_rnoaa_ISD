```python
import numpy             as np
import pandas            as pd
import xarray            as xr
import matplotlib.pyplot as plt
import datetime          as datetime
```

```python
nCLIMDIV_url = "http://kyrill.ias.sdsmt.edu:8080/thredds/fileServer/CLASS_Examples/nCLIMDIV.csv"

ftyates1_url      = "http://kyrill.ias.sdsmt.edu:8080/thredds/dodsC/GHCNd/USA/ND/netCDF/GHCND-USC00323206__FORT_YATES_ND_US.nc"
ftyates2_url      = "http://kyrill.ias.sdsmt.edu:8080/thredds/dodsC/GHCNd/USA/ND/netCDF/GHCND-USC00323207__FORT_YATES_4_SW_ND_US.nc"


nCLIMDIV     = pd.read_csv(filepath_or_buffer=nCLIMDIV_url)
ftyates1     = xr.open_dataset(filename_or_obj=ftyates1_url)
ftyates2     = xr.open_dataset(filename_or_obj=ftyates2_url)

display(ftyates1)
display(ftyates2)
```

```python
ftyates1 = ftyates1["precipitation_amount"].to_pandas()
ftyates1['Year'] = pd.to_datetime(ftyates1.index).year



ftyates2 = ftyates2["precipitation_amount"].to_pandas()

```

```python
ftyates1["precipitation_amount"].plot()
ftyates2["precipitation_amount"].plot()
```

```python
nCLIMDIV = nCLIMDIV[nCLIMDIV["Full_Zone_Code"] == 3208]
nCLIMDIV['Year'] = pd.to_datetime(nCLIMDIV['Date']).dt.year
nCLIMDIV = nCLIMDIV[nCLIMDIV["Year"] < 2025]



nCLIMDIV_annual = nCLIMDIV.                  \
    groupby('Year').                         \
       agg(nCLIMDIV_prec = ('PCPN', 'sum')). \
    reset_index().set_index("Year")

nCLIMDIV_annual.plot()
```
